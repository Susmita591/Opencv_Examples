```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files
uploaded = files.upload()
```

```python
filename = list(uploaded.keys())[0]
image = cv2.imread(filename)
cv2_imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resized = cv2.resize(image, [300,300])
cv2_imshow(resized)
```

```python
a=[image,gray]
for i in a:
  resized = cv2.resize(i,(300, 300))
  cv2_imshow(resized)
```

```python
# draw rectangle and text
# draw rectangle and text
annotated = image.copy()
cv2.rectangle(annotated,(50,50),(250,250),(0,0,255),3)
cv2.putText(annotated, 'My Box', (60,45), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0))
# Use cv2_imshow instead of cv2.imshow
from google.colab.patches import cv2_imshow # Importing the necessary function
cv2_imshow(annotated) # Displaying the image using cv2_imshow
```

```python
cv2.imwrite('output_image.jpg', annotated)
print("Image saved as output_image.jpg")
```
